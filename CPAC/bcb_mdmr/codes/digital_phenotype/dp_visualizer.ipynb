```python
import pandas as pd
import seaborn as sns

group = "dp_call"
variable = "phone_calls_rapids_incoming_maxduration_weekdays"

regressor = pd.read_csv(f"../../input/{group}_{variable}_regressor.csv")
regressor['group'] = regressor['Participant'].apply(lambda x: 'HC' if x.startswith('c') else 'SAD')
regressor.columns

LSAS_regressor = pd.read_csv(f"../../input/gangnam_total_LSAS_regressor.csv")
HADS_depression_regressor = pd.read_csv(f"../../input/gangnam_total_HADS_depression_regressor.csv")
```

```python
sns.displot(
    regressor, 
    x=variable, hue="group",
)
```

```python
sns.displot(
    regressor, 
    x=variable, hue="group",
    kind="kde"
)
```

```python
dp_LSAS = pd.merge(regressor, LSAS_regressor, on='Participant', how='left')
dp_HADS_depression = pd.merge(regressor, HADS_depression_regressor, on='Participant', how='left')
```

```python
sns.relplot(data=dp_LSAS, x=variable, y="LSAS", hue="group")
```

```python
sns.relplot(data=dp_HADS_depression, x=variable, y="HADS_depression", hue="group")
```
