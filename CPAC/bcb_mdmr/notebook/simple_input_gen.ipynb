```python
import pandas as pd
import os

group_name = "gangnam_1"

predictor_file_name = "../input/participant_demo_clinical_all.csv"

variable_of_interest_list = ['LSAS', 'LSAS_performance', 'LSAS_social_interaction', 'BFNE']

columns_of_interest = ['1. SEX', '2.AGE', '3-2. YR_EDU', 'fmri_code', 'LSAS', 
                       'LSAS_performance', 'LSAS_social_interaction', 'BFNE']


predictor = pd.read_csv(predictor_file_name)
predictor
```

```python
cleaned_predictor = predictor.dropna(subset=columns_of_interest)
cleaned_predictor
```

```python
import os
import re
fmriprep_dir = "/mnt/NAS2/data/SAD_gangnam_resting/fMRIPrep/"
# List all files and directories in the given path
items = os.listdir(fmriprep_dir)

# Define a regex pattern to match the codes
pattern = re.compile(r'sub-(c\d{4}|s\d{4})')

# Extract unique codes
unique_codes = set()
for item in items:
    match = pattern.match(item)
    if match:
        unique_codes.add(match.group(1))

# Convert the set to a sorted list
unique_codes = sorted(list(unique_codes))
```

```python
# Filter the cleaned_predictor dataframe to include only rows with fmri_code matching unique_codes
filtered_predictor = cleaned_predictor[cleaned_predictor['fmri_code'].isin(unique_codes)]
filtered_predictor
```

```python
participant_df = filtered_predictor.rename(columns={
    'fmri_code': 'Participant',
    '1. SEX': 'SEX',
    '2.AGE': 'AGE',
    '3-2. YR_EDU': 'YR_EDU'
})[['Participant', 'SEX', 'AGE', 'YR_EDU', 'LSAS', 'LSAS_performance', 'LSAS_social_interaction', 'BFNE']]
participant_df['SEX'] = participant_df['SEX'].map({1: 1, 2: 0})
participant_df
```

```python
def save_variable_of_interest(df, variables):
    for variable in variables:
        # Select the necessary columns
        df_subset = df[['Participant', 'SEX', 'AGE', 'YR_EDU', variable]]
        
        # Create the filename
        base_dir = "/home/changbae/fmri_project/C-PAC/CPAC/bcb_mdmr/input/"
        filename = f"{base_dir}gangnam_1_{variable}_regressor.csv"
        
        # Save the DataFrame to a CSV file
        df_subset.to_csv(filename, index=False)
        print(f"Saved {filename}")
```

```python
save_variable_of_interest(participant_df,variable_of_interest_list)
```

```python
# Save the 'Participant' column to a separate CSV file
participant_codes = participant_df[['Participant']]

# Save the DataFrame to a CSV file
participant_codes.to_csv("/home/changbae/fmri_project/C-PAC/CPAC/bcb_mdmr/input/gangnam_1_code_list.csv", index=False, header=False)
print("Saved participant codes")
```
